### Apply zonal statistics on GEDI data with Lidar data

```python
import glob
from rasterio.merge import merge
from rasterio.plot import show
import rasterio
```

#### Read Lidar data

```python
## Read in airborne Lidar (CHM)

lidar_string = glob.glob("../../../data/CHM/Kituntu*.tif")

# Select lidar tiles in Area (Up)
lidarUp = lidar_string[0:8]
print("\n".join(lidarUp),"\n") # tif files belonging to area (Up)
```

```python
## Merge lidarUp 

mosaic, output = merge(lidarUp)
show(mosaic)

# Add metadata (with dimensions and crs etc)
with rasterio.open(lidarUp[1]) as src:
        out_meta = src.meta.copy()   

out_meta.update({"driver": "GTiff",
                 "height": mosaic.shape[1],
                 "width": mosaic.shape[2],
                 "transform": output})

# Save merged tif file
with rasterio.open("../../../data/processed/lidarUpMerged.tif", 'w', **out_meta) as f:
    f.write(mosaic)
```

```python
# Function to reproject CRS of raster file

def reproject_raster(in_path, out_path, crs):

    # reproject raster to crs
    with rasterio.open(in_path) as src:
        src_crs = src.crs
        transform, width, height = rasterio.warp.calculate_default_transform(src_crs, crs, src.width, src.height, *src.bounds)
        kwargs = src.meta.copy()

        kwargs.update({
            'crs': crs,
            'transform': transform,
            'width': width,
            'height': height})

        with rasterio.open(out_path, 'w', **kwargs) as dst:
            for i in range(1, src.count + 1):
                rasterio.warp.reproject(
                    source=rasterio.band(src, i),
                    destination=rasterio.band(dst, i),
                    src_transform=src.transform,
                    src_crs=src.crs,
                    dst_transform=transform,
                    dst_crs=crs,
                    resampling=rasterio.warp.Resampling.nearest)
    return(out_path)
```

```python
# Reproject lidar rasters to CRS of training plots

#trainPolygons.crs
crs = "EPSG:32624"
crs = gdf.crs

# Change crs of area Up
in_path = "../../data/processed/lidarUpMerged.tif"
out_path = "../../data/processed/lidarUpMergedGEDIcrsTest.tif"
reproject_raster(in_path, out_path, crs) # autocreates tif file with new crs
```

#### Zonal statistics function

```python
from rasterstats import zonal_stats
```

```python
# NaNs might appear due to pixels not in the extent of the lidar Area.
```

```python
# Statistics to compute
stats=['min','max','mean','std','sum', 'percentile_10',
       'percentile_20','percentile_30','percentile_40',
       'percentile_50','percentile_60','percentile_70',
       'percentile_75','percentile_80','percentile_85',
       'percentile_90','percentile_95','percentile_98']

# Zonal stats
zon = gdf.join(
    pd.DataFrame(
        zonal_stats(
            vectors=gdf['geometry'], 
            raster="../../data/processed/lidarUpMergedGEDIcrsTest.tif", 
            stats=stats
        )
    ),
    how='left'
)
zon
```

```python
ax = zon[zon['mean'].isnull()].plot()
#temp.plot()
#gdf.plot()
# temp.loc[0:150].plot(color="red")
# plt.show()
# show(temp, ax=ax)
```

```python
# Count total of NaNs per column 

zon[[*(stats)]].isna().sum()
#temp.isna().sum()
```

```python
# Remove all NaN pixels

zon = zon[zon['min'].notna()]
zon[[*(stats)]].isna().sum()
zon
```

```python
# Check if all nans are removed
zon.isna().sum()
```
